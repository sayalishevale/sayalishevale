```python
# Locations

location_resp = requests.get(data['locations'])
location_data = location_resp.json()
location_df = pd.DataFrame(location_data['results'])
location_df.head()
```

```python
# Episodes

episode_resp = requests.get(data['episodes'])
episode_data = episode_resp.json()
episode_df = pd.DataFrame(episode_data['results'])
episode_df.head()
```

```python
# List the names of all characters present in an episode released in 2014.

# Map URLs to character names, ensuring there are matches before accessing data.
characters_names_2014 = []
for url in characters_names_2014:
  matching_characters = character_df[character_df['url'] == url]
  if not matching_characters.empty:
    character_name = matching_characters['name'].iloc[0]
    characters_names_2014.append(character_name)
    print(characters_names_2014)
  else:
     print(f"No matching records found for URL: {url}")
```

```python
# Name the last location of Annie, from most recent episode.

#Find the character entry for Annie
annie_url = character_df[character_df['name'] == 'Annie']['url'].iloc[0]

#find episode featuring Annie
annie_episodes = episode_df[episode_df['characters'].apply(lambda x: annie_url in x)]

# If the episodes are not in chronological order, sort them by air_date.
annie_episodes = annie_episodes.sort_values(by = 'air_date',ascending = False)

# Get the most recent episode's information.
most_recent_episode = annie_episodes.iloc[0]

# Assuming that 'location' contains URLs, match the URL to an entry within the location DataFrame.
last_seen_location_url = character_df[character_df['url'] == annie_url]['location'].iloc[0]['url']
last_seen_location_name = location_df[location_df['url'] == last_seen_location_url]['name'].iloc[0]
last_seen_location_name
```

```python
# List the names of all episodes in which both Jerry Smith and Summer Smith appeared.

Jerry_and_Summer_episodes = episode_df[episode_df['characters'].apply(lambda x: 'character/3' in x and 'character/5' in x)]
# Extract the names of this eppisode
Jerry_and_Summer_episode_names = Jerry_and_Summer_episodes['name'].tolist()
Jerry_and_Summer_episode_names
```

```python
# List the names of all characters who appeared in an episode in which Amish Cyborg also appeared.
amish_cyborg_episodes = episode_df[episode_df['characters'].apply(lambda x: "character/{Amish_Cyborg_id}" in x)]
amish_cyborg_episodes

characters_in_amish_cyborg_episodes = amish_cyborg_episodes['characters']


other_characters = [url for url in characters_in_amish_cyborg_episodes if url != "character/{Amish_Cyborg_id}"]
other_character_names = [character_df[character_df['url'] == url]['name'].iloc[0] for url in other_characters]
other_character_names

```

```python
# List top 5 characters with maximum number of episodes. Start from the highest.

character_appearance_counts = episode_df['characters'].explode().value_counts()
top_characters = character_appearance_counts.head(5)

top_character_names = [character_df[character_df['url'] == url]['name'].iloc[0] for url in top_characters.index]
top_character_names
```
