```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.imdb.com/search/title/?groups=top_1000&view=simple&sort=user_rating,desc&count=100"

# Send an HTTP GET request to the URL
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    movie_containers = soup.find_all("div", class_="lister-item-content")

    movie_data = []

    for movie in movie_containers:
        title = movie.find("a", href=True).text
        year = movie.find("span", class_="lister-item-year").text.strip("()")

        movie_data.append({"Title": title,"Year": year})

    # Create a DataFrame to store the data
    df = pd.DataFrame(movie_data)

    # Display the DataFrame
    print(df)

else:
    print("Failed to retrieve the page. Status code:", response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup

# Send an HTTP GET request to the IMDb URL
url = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')

    movie_containers = soup.find_all("div", class_="lister-item-content")

    for movie in movie_containers:
        # Movie Title
        title = movie.find("h3").find("a").text

        # Release Year
        year = movie.find("span", class_="lister-item-year").text.strip("()")

        # Movie Span (e.g., PG-13)
        span = movie.find("span", class_="certificate").text if movie.find("span", class_="certificate") else "Not Rated"

        # IMDb Rating
        rating = movie.find("div", class_="inline-block ratings-imdb-rating").text.strip()

        # Gross (if available)
        gross = movie.find("p", class_="sort-num_votes-visible").find_all("span")[-1].text if movie.find("p", class_="sort-num_votes-visible") else "N/A"

        # Director (if available)
        director = movie.find("p", class_="").find_all("a")[-1].text if movie.find("p", class_="").find_all("a") else "N/A"

        print("Title:", title)
        print("Year:", year)
        print("Span:", span)
        print("Rating:", rating)
        print("Gross:", gross)
        print("Director:", director)
        print("-" * 40)
else:
    print("Failed to retrieve the page. Status code:", response.status_code)


```

```python
movie_df = pd.DataFrame(movie_data, columns=['title','year','span','rating','gross','director'])
movie_df.head(10)
```



*   Title: Extracted from an anchor <a> element nested inside an <h3> element.
*   Year: Extracted from a <span> element with the class "lister-item-year." The .strip("()") method removes parentheses from around the year.


*   Span: Extracted from a <span> element with the class "certificate." If not present, "Not Rated" is used as the default.

*   Rating: Extracted from a <div> element with the class "inline-block ratings-imdb-rating."

*   Gross: Extracted from a paragraph <p> element with the class "sort-num_votes-visible." The last <span> element within the paragraph typically contains gross information. If not available, "N/A" is used.


*   Director: Extracted from a paragraph <p> element with no specific class, which contains a list of links. The last link within this paragraph is assumed to be the director's name. If not available, "N/A" is used as the default.
