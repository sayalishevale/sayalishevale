```python
import pandas as pd

character_df = pd.DataFrame(character_data['results'])
character_df.head()
```

```python
## Locations
location_resp = requests.get(data['locations'])
location_data = location_resp.json()
location_df = pd.DataFrame(location_data['results'])
location_df.head()
```

```python
## Episodes
episode_resp = requests.get(data['episodes'])
episode_data = episode_resp.json()
episode_df = pd.DataFrame(episode_data['results'])
episode_df.head()
```

```python
# List the names of all the episodes released (air_date) in 2014.

# Convert 'air_date' from string to datetime for easy comparison
episode_df['air_date'] = pd.to_datetime(episode_df['air_date'])

# Filter episodes released in 2014
episodes_2014 = episode_df[episode_df['air_date'].dt.year == 2014]

# Display the names of the episodes
print(episodes_2014['name'].tolist())
```

```python
# What is the origin location of Rick Sanchez ?

rick_origin_url = character_df[character_df['name'] == 'Rick Sanchez']['origin'].iloc[0]['url']

# Make a new request to fetch Rick's origin location details
rick_origin = requests.get(rick_origin_url).json()

print(rick_origin['name'])
```

```python
# List the names of all the episodes with a release date on or after 2014, in alphabetical order.

# Filter episodes based on air_date and sort them
episodes_2014_onwards = episode_df[episode_df['air_date'].dt.year >= 2014].sort_values(by='name')

# Display the sorted list of episode names
print(episodes_2014_onwards['name'].tolist())
```

```python
# List the names and release years of all the episodes where Summer Smith is present, in chronological order.

# Identify Summer Smith's URL from the characters data
summer_url = character_df[character_df['name'] == 'Summer Smith']['url'].iloc[0]

# Find episodes where Summer Smith is present
summer_episodes = episode_df[episode_df['characters'].apply(lambda x: summer_url in x)]

# Extract the year and sort the data chronologically
summer_episodes['year'] = summer_episodes['air_date'].dt.year
summer_episodes = summer_episodes.sort_values(by='air_date')[['name', 'year']]

print(summer_episodes)
```

```python
# What are the total episodes in which Jerry Smith appeared?

jerry_episodes = episode_df[episode_df['characters'].apply(lambda x: any("character/5" in s for s in x))]
total_jerry_episodes = len(jerry_episodes)
```

```python
# List all characters created in 2017, in descending order by species. For characters with the same species, order them alphabetically by their name.

characters_2017 = character_df[pd.to_datetime(character_df['created']).dt.year == 2017]
sorted_characters = characters_2017.sort_values(by=['species', 'name'], ascending=[False, True])
sorted_characters
```

```python
# List the names of all characters present in S01E05.

characters_in_s01e05 = episode_df[episode_df['episode'] == 'S01E05']['characters'].iloc[0]
# You need to map these character URLs back to their names using the character_df DataFrame.
```

```python
# Map URLs to character names, ensuring there are matches before accessing data.
characters_names_2014 = []
for url in characters_names_2014:
    matching_characters = character_df[character_df['url'] == url]
    if not matching_characters.empty:
        character_name = matching_characters['name'].iloc[0]
        characters_names_2014.append(character_name)
        print(characters_names_2014)
    else:
        print(f"No matching records found for URL: {url}")
```

```python
# Name the last location of Annie, from the most recent episode.

# Find the character entry for Annie.
annie_url = character_df[character_df['name'] == 'Annie']['url'].iloc[0]

# Find episodes featuring Annie.
annie_episodes = episode_df[episode_df['characters'].apply(lambda x: annie_url in x)]

# If the episodes are not in chronological order, sort them by air_date.
annie_episodes = annie_episodes.sort_values(by='air_date', ascending=False)

# Get the most recent episode's information.
most_recent_episode = annie_episodes.iloc[0]

# Assuming that 'location' contains URLs, match the URL to an entry within the location DataFrame.
last_seen_location_url = character_df[character_df['url'] == annie_url]['location'].iloc[0]['url']
last_seen_location_name = location_df[location_df['url'] == last_seen_location_url]['name'].iloc[0]
last_seen_location_name
```

```python
# List the names of all episodes in which both Jerry Smith and Summer Smith appeared.

# Identify episodes where both characters appear.
jerry_and_summer_episodes = episode_df[
    episode_df['characters'].apply(lambda x: "character/3" in x and "character/5" in x)
]

# Extract the names of these episodes.
jerry_and_summer_episode_names = jerry_and_summer_episodes['name'].tolist()
jerry_and_summer_episode_names
```

```python
# List the names of all characters who appeared in an episode in which Amish Cyborg also appeared.

# Find the episodes where Amish Cyborg appears.
amish_cyborg_episodes = episode_df[
    episode_df['characters'].apply(lambda x: "character/{Amish_Cyborg_id}" in x)]  # replace {Amish_Cyborg_id} with the character's actual ID
amish_cyborg_episodes

# Get all characters from these episodes.
characters_in_amish_cyborg_episodes = amish_cyborg_episodes['characters']

# Exclude Amish Cyborg from the list and map URLs to character names.
other_characters = [url for url in characters_in_amish_cyborg_episodes if url != "character/{Amish_Cyborg_id}"]
other_character_names = [character_df[character_df['url'] == url]['name'].iloc[0] for url in other_characters]
other_character_names
```

```python
# List top 5 characters with the maximum number of episodes. Start from the highest.

# Count the number of appearances by character.
character_appearance_counts = episode_df['characters'].explode().value_counts()

# Get the top 5 characters.
top_characters = character_appearance_counts.head(5)

# Map the character URLs to names.
top_character_names = [character_df[character_df['url'] == url]['name'].iloc[0] for url in top_characters.index]
top_character_names
```
